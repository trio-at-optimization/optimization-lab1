```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from decimal import *

import pandas as pd
import scipy
from matplotlib import pylab, gridspec, pyplot as plt

%matplotlib inline
plt.style.use('fivethirtyeight')

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d
```

```python
def print_full_grad(result, label = 'градиент', elev = 50, azim = 100):
  fig = plt.figure(figsize=(10, 10))
  ax = fig.add_subplot(projection='3d')
  ax.plot(xs=result[:, 0], ys=result[:, 1], zs=f(np.stack((result[:, 0], result[:, 1]))), marker='', markersize=25,
                  markerfacecolor='y', lw=3, c='black', label=label)

  ax.plot_surface(X, Y, f(np.stack((X, Y))), cmap=cm.coolwarm)
  ax.view_init(elev=elev, azim=azim)
  ax.legend(loc='upper left')
  print(result[-1], f(result[-1]))
  plt.show()
```

```python
def f(x):
    return x[0] ** 2 + (x[0] - x[1]) ** 2

def nabla(x):
    return [(2 * x[0] + 2 * (x[0] - x[1])), (-2 * (x[0] - x[1]))]
```

```python
x = np.linspace(-3, 7, 120)
X, Y = np.meshgrid(x, x)
```

```python
points = np.array([[5.0, 5.0]])
alpha = 0.001
k = 10000

for i in range(1, k + 1):
    a = nabla(points[i - 1])[0]
    b = nabla(points[i - 1])[1]
    c = points[i - 1, 0] - a * alpha
    d = points[i - 1, 1] - b * alpha
    points = np.vstack([points, [c, d]])

print(points[-1], f(points[-1]))
```

```python
print_full_grad(points, elev=20, azim=-80)
```
